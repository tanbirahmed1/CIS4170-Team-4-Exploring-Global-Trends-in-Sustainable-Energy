**We wanted to analyze which countries have the highest Renewable Energy Capacity in the world**

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("global-data-on-sustainable-energy.csv")


country_production = data.groupby("Entity")["Electricity from renewables (TWh)"].sum().reset_index()


top_countries = country_production.sort_values(by="Electricity from renewables (TWh)", ascending=False).head(10)


plt.figure(figsize=(12, 6))
plt.bar(top_countries["Entity"], top_countries["Electricity from renewables (TWh)"], color='b')
plt.title("Top 10 Countries with the Highest Electricty from Renewables (2000-2020)")
plt.xlabel("Country")
plt.ylabel("Total Renewable Capacity (Megawatts)")
plt.xticks(rotation=45, ha='right')
plt.grid(True)


plt.tight_layout()
plt.show()
```

**We wanted to visualize the Global Sustaianble Energy Production Trends from (2000-2020)**

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("global-data-on-sustainable-energy.csv")


global_production = data.groupby("Year")["Access to electricity (% of population)"].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(global_production["Year"], global_production["Access to electricity (% of population)"], marker='o', linestyle='-')
plt.title("Global Sustainable Energy Production Trend (2000-2020)")
plt.xlabel("Year")
plt.ylabel("Access to Electricity")
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv("global-data-on-sustainable-energy.csv")


import seaborn as sns

# Filter the data for the relevant columns
relevant_columns = ["Entity", "Year", "gdp_per_capita", "Renewable energy share in the total final energy consumption (%)"]
filtered_data = data[relevant_columns]

# Pivot the data to create a pivot table for the heatmap
heatmap_data = filtered_data.pivot(index="Entity", columns="Year", values="gdp_per_capita")

# Create a customized heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data.corr(), annot=True, fmt=".2f", cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Heatmap: GDP per Capita vs. Renewable Energy Share")
plt.xlabel("Year")
plt.ylabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

**We wanted to analyze the Top 10 countries where thier percentage of final consumption of energy is derived from renewable resources.**

```python
import matplotlib.pyplot as plt


relevant_columns = ["Entity", "Renewable energy share in the total final energy consumption (%)"]
filtered_data = data[relevant_columns]


average_renewable_share = filtered_data.groupby("Entity")["Renewable energy share in the total final energy consumption (%)"].mean()


top_10_countries = average_renewable_share.nlargest(10)


plt.figure(figsize=(12, 6))
top_10_countries.plot(kind="bar", color="skyblue")
plt.title("Top 10 Countries with the Highest Average Renewable Energy Share")
plt.xlabel("Country")
plt.ylabel("Average Renewable Energy Share (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
